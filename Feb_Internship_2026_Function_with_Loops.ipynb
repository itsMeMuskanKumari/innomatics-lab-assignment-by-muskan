Problem 1: Mobile Recharge Validation

```python
def validate_recharge(amount, valid_plans):
    if amount < 50:
        return "Recharge amount must be at least ₹50."
    elif amount not in valid_plans:
        return "Invalid plan selected."
    else:
        return "Recharge Successful!"

valid_plans = [199, 299, 399, 599]

while True:
    try:
        amount = int(input("Enter recharge amount: ₹"))
        result = validate_recharge(amount, valid_plans)
        print(result)

        if result == "Recharge Successful!":
            break
        else:
            print("Please try again.\n")
    except ValueError:
        print("Invalid input. Please enter a valid number.\n")
```

Problem 2: Inventory Reorder Alert System

```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

inventory = {
    "Laptop": 10,
    "Mouse": 25,
    "Keyboard": 12,
    "Monitor": 18
}

check_inventory(inventory)
```

Problem 3: Student Result Processing System

```python
def process_result(marks):
    total = 0
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average: {average:.2f} - Pass"
    else:
        return f"Average: {average:.2f} - Fail"

student_marks = [60, 45, 70, 55]

result = process_result(student_marks)
print(result)
```

Problem 4: Cab Fare Estimator with Retry Option

```python
def calculate_fare(distance, peak_hour):
    base_fare = 50
    fare = base_fare + (12 * distance)

    if peak_hour.lower() == "yes":
        fare += fare * 0.25   # 25% extra

    return fare

while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak)
        print(f"Total Fare: ₹{total_fare:.2f}")

        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            break
    except ValueError:
        print("Invalid input. Please try again.\n")
```

Problem 5: Employee Attendance Eligibility Checker

```python
def check_attendance(attendance_list):
    present_days = 0

    for status in attendance_list:
        if status.upper() == "P":
            present_days += 1

    percentage = (present_days / len(attendance_list)) * 100

    if percentage >= 75:
        return f"Attendance: {percentage:.2f}% - Eligible"
    else:
        return f"Attendance: {percentage:.2f}% - Not Eligible"

attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

result = check_attendance(attendance)
print(result)
```

Problem 6: Password Strength Checker

```python
def check_password_strength(password):
    if len(password) < 8:
        return "Weak Password: Must be at least 8 characters long."

    has_digit = False
    has_special = False
    special_chars = "@#$"

    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True

    if not has_digit:
        return "Weak Password: Must contain at least one digit."
    if not has_special:
        return "Weak Password: Must contain at least one special character (@#$)."

    return "Strong Password!"

password = input("Enter password: ")
print(check_password_strength(password))
```
